## **Decision Tree**

A Decision Tree is a machine learning algorithm used for classification and regression. It splits the dataset into branches based on feature values to make predictions.

* **Root Node:** The starting point (first decision) of the tree
* **Decision Node:** Internal nodes where the tree splits based on a condition
* **Leaf Node:** The final outcome (prediction)
* **Splitting:** Dividing data based on a feature
* ID3 (C4.5): entropy || Cart: Gini (default) [algorithm used in Decision Tree]
* **Gini Impurity:** Measures how impure a node is (how mixed the classes are)
** `Gini=1 − ∑pi**2​`
** where 𝑝i is the probability of class i in the node.

*  **Entropy & Information Gain:**
** **Entropy** measures disorder (uncertainty)- how mixed the data is.
** High Entropy = Very mixed, Low Entropy = Mostly one class, Entropy = 0 means a pure node
** `Entropy = −∑pi log2(pi)`
** **Information Gain** measures how much entropy is reduced after a split
** `IG = Entropy before − Entropy after`
** higher IG = better split

* **Pruning:** Pruning removes unnecessary branches to make the tree simpler and prevent overfitting.
** **reduce entropy = improve Information Gain**
** Types of Pruning:
** 1. Pre-Pruning (Early Stopping): Stop growing the tree if a split does not improve Information Gain much (reduce entropy).
** 2. Post-Pruning (After Training): Grow the full tree, then remove weak branches. If a branch only applies to 1% of data, remove it.

* **Overfitting:** When a tree is too deep, memorizing training data instead of learning patterns

* **How to Choose the Root Node in a Decision Tree?**
The root node is the best feature to split the dataset first. We choose it based on:
** Calculate Gini or Entropy for Each Feature
** The feature with the Lowest Gini or Highest Information Gain will becomes the Root Node

## **Random Forest**
**Random Forest** is an **ensemble learning** method that combines multiple Decision Trees using **Bagging** (Bootstrap Aggregating) to improve accuracy and reduce overfitting.

**How It Works**
* Creates multiple Decision Trees -
** Bootstrap Sampling → Randomly select different subsets of the data (rows) with replacement. Some rows may appear multiple times, and some may not be selected at all.
** Feature Randomness → Instead of using all features, each tree selects a random subset of features (columns) at every split.
** Build a decision tree for each subset.

* Train Each Tree Independently
** Each decision tree learns different patterns because of randomness.

* Final output
** Combines all tree predictions for better accuracy.
** Majority vote (for classification) or Average (for regression).

**Bagged-Tree** also use bagging but without feature randomness.
* Random Forest builds upon Bagged Trees by adding Feature Randomness  
`Random Forest = Bagged Trees + Feature Randomness`

## **Support Vector Machine (SVM)**
SVM is a supervised learning algorithm used for classification and regression. It finds the best decision boundary (hyperplane) that separates different classes in a dataset.

* Find the Best Hyperplane
** The hyperplane is the decision boundary that separates classes.
SVM tries to maximize the margin between classes.

* Support Vectors
** The closest data points to the hyperplane are called support vectors.
These define the margin and influence the position of the hyperplane.


**C (Regularization):** Controls how much SVM tries to avoid misclassification (tries to classify evry point correctly)
* High C = tries to classify everything correctly. (strict)
* Low C = allows some mistakes. (relaxed)

**Gamma (γ):** SVM controls how much influence a single training point has on the decision boundary.
* High γ = focuses on nearby points.
* Low γ = considers distant points too.

**Hyperplane Selection**
* **Linear SVM:** If data is linearly seperable, SVM finds a straight line or plane in higher dimension.
* **Non-linear SVM:** If data is not seperable by a straight line, we use Kernels to transform the data.

**Kernel:** It is a mathematical function that helps organize data and make it easier to classify.

* **Kernel Trick:** If data is not linearly separable, SVM uses kernels to map it to a higher dimension where it becomes separable.
** Linear Kernel → When data is already separable.
** Polynomial Kernel → When relationships between data points are more complex.
** RBF ( Radial Basis Function) (Gaussian) Kernel → When the decision boundary is highly non-linear.
** Sigmoid Kernel → When mimicking a neural network approach.

## **Regularization:**
* Regularization prevents overfitting by making the model simpler so it works well on new data. It adds a penalty to large weights (coefficients) in the model.

**Types of Regularization**
* L1 Regularization (Lasso) → Makes some weights zero (removes less important features).
** Example: Choosing only the most important topics to study instead of everything.

* L2 Regularization (Ridge) → Makes weights smaller but doesn’t remove them completely.
** Example: Giving less importance to extra details but not ignoring them.

### **Lasso Regression**
* Lasso Regression (Least Absolute Shrinkage and Selection Operator) is a linear regression technique that uses L1 regularization to reduce overfitting and select important features.

* Lasso adds a **penalty** to the loss function, forcing some weights to become exactly zero. This helps in:
** --> Reducing overfitting (simplifies the model).
** --> Feature selection (removes less important features).

**Lasso Loss Function (with L1 Regularization)**

**`Loss = ∑(y - y!)^2 + λ ∑ ∣w∣`**

∑(y − y^) 2  → Measures prediction error.

λ ∑ ∣w∣ → L1 penalty (shrinks some weights to zero).

λ → Controls how much regularization is applied.

###**Ridge Regression**
* Ridge Regression is a type of linear regression that uses L2 regularization to prevent overfitting.

* Ridge adds a penalty to the loss function, which shrinks the weights but never makes them exactly zero. This helps in:
** --> Reducing overfitting (prevents large weights).
** --> Handling multicollinearity (when features are highly correlated).

**Ridge Loss Function (with L2 Regularization)**

**`Loss = ∑(y − y!)^2 + λ ∑ w^2`**

∑(y − y!)^2 → Measures prediction error.

λ ∑ w^2 → L2 penalty (shrinks weights).

λ → Controls how much regularization is applied.

## **Naïve Bayes**
Naïve Bayes is a classification algorithm based on probabilities. It is called "Naïve" because it assumes all features are independent, even if they are not in reality.

* Convert Data into Probabilities → Calculate how often each class and feature appears.
* Apply Bayes' Theorem → Predict the class with the highest probability.

**`P(Class∣Data) = P(Data∣Class) × P(Class) / P(Data)`**

P(Class∣Data) = Probability of class given the data (we want to find this).
P(Data∣Class) = Likelihood (how likely data belongs to class).
​P(Class) = Prior probability of the class.
P(Data) = Probability of the data (same for all classes, so we ignore it).

**Types of Naïve Bayes:**
* Gaussian Naïve Bayes → For continuous data (e.g., height, weight).
* Multinomial Naïve Bayes → For text classification (e.g., spam detection).
* Bernoulli Naïve Bayes → For binary features (e.g., word presence/absence in an email).

**Works Well with Small Data → Needs fewer training examples.**

## **K-Nearest Neighbors (KNN)**
**K-Nearest Neighbors (KNN)** is a **non-parametric**, **supervised** learning algorithm used for **classification** and **regression**. It predicts the class of a new data point based on the **majority class of its K nearest neighbors.**

* Choose a value for K (number of neighbors to consider).
* Find the closest K (neighbors) to the new data point using a distance metric (eg. Euclidean distance).
* For classification: Assign the most common (majority) class among the K neighbors.
* For regression: Take the average of the K neighbors' values.

**Example:**

**Imagine we want to classify a new data point (🔴)**

K = 3 (3 nearest neighbors considered):

2 🟢 (Green class)
1 🔵 (Blue class)

📌 Since the majority class is Green (🟢), the new point is classified as Green.


**Euclidean Distance (Most Common Distance Used):** This measures how close two points are in space.

* **`d = sqrt( (x2 − x1)^2 + (y2 − y1)^2 )`**

## **Boosting**

* **Gradient Boosting** is an **ensemble learning** method that combines multiple weak models (usually decision trees) to create a strong model.
* Models are trained **sequentially** (one after another), where each new model(tree) **corrects the errors** of the previous ones.

**Types of Boosting Algorithms**
* **AdaBoost (Adaptive Boosting)** – Adjusts weights of misclassified points. (Weight-based Boosting)
* **Gradient Boosting Machine (GBM)** – Uses gradient descent to minimize errors. (Error-based Boosting)
** **XGBoost** (Optimized GBM) → Faster, regularized  
** **LightGBM** (Efficient GBM) → Leaf-wise growth, fast on large data  
** **CatBoost** (Categorical GBM) → Best for categorical data

### **Gradient Boosting:**  
**Gradient Boosting** is a specific type of Boosting that uses **gradient descent.**

* **Start with a weak model** (like a small decision tree).
* **Find the errors** (residuals) in the model’s predictions.
* **Train a new tree** to correct those errors.
* **Repeat the process** until the model improves.
* **Combine all trees** to make the final prediction.

**Gradient Boosting- XGBoost, LightGBM, and CatBoost**  Histogram-Based, Stochastic

* **XGBoost** (Extreme Gradient Boosting): Fast and optimized version
** Uses regularization (L1 & L2) to reduce overfitting.
** Handles missing values automatically.
** Uses parallel computation (faster than basic Gradient Boosting).
** Best for: General-purpose boosting, large datasets.

* **LightGBM** (Light Gradient Boosting Machine): Faster for large datasets.
** Splits trees leaf-wise (instead of level-wise) → faster & more accurate.
** Handles large datasets efficiently.
** Needs less memory than XGBoost.
** Best for: Large datasets, real-time predictions.

* **CatBoost** (Categorical Boosting): Works well with categorical data.
** Handles categorical data directly (no need to convert to numbers).
** Works well with imbalanced data.
** Fast training speed.
** Best for: Datasets with many categorical features (e.g., customer data, medical records).

### **K-Means**
Groups similar data points into **K** clusters.
* Choose how many clusters (K) you need.
* Pick K random points as the starting cluster centers (centroids).
* Assign each data point to the nearest centroid.
* Update centroids by taking the average of all points in each cluster.
* Repeat steps 3-4 until the centroids stop moving (they stay the same).

k-means.png

**Choosing the Right K (Elbow Method)**

The Elbow Method uses SSE (**Sum of Squared Errors**). It checks how much the **error reduces** as the number of **clusters increases**.

* When we increase the number of clusters (K), the error (SSE) decreases. This means the clustering is improving.
* Initially, SSE decreases rapidly (e.g., from K=1 to K=2 to K=3).
* After a certain point, SSE decreases very slowly.
* This sharply bending point is called the "Elbow Point", and it represents the optimal K value.
* After this point, adding more clusters does not improve clustering but increases complexity.

elbow-method.png

**If the new centroids are equal to the old centroids, then the algorithm has converged!**
* **`New centroids = Old centroids`**

### **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**
**DBSCAN** groups data points **based on density** and can detect **noise** (outliers). Unlike K-Means, it doesn’t need to predefine the number of clusters (K) and works well with irregularly shaped clusters.

* Define ε (neighborhood distance) and MinPts (minimum points to form a cluster).  
* Pick a random point:  
    - If it has at least MinPts neighbors → Form a cluster.  
    - Else, mark it as noise (for now).  
* Expand the cluster by adding neighboring points.  
* Repeat for new points until the cluster stops growing.  
* Mark remaining unclustered points as noise (outliers).  
* Repeat until all points are processed.

dbscan.png

* **Define Two Parameters:**
** **Epsilon (ε) –** The maximum distance to consider points as neighbors.
** **MinPts (Minimum Points) –** Minimum number of points required to form a dense region (cluster).

If **MinPts = 4**, a point needs at least **4 neighbors** within ε distance to form a cluster.

If a point has fewer than 4 neighbors, it’s either border or noise.


* **Pick Any Random Point (unvisited)**
** If this point has at least MinPts = n  neighbors within ε, it becomes a core point and starts forming a cluster.
** If it has fewer than MinPts = n  neighbors, it remains noise (temporarily).


* **Expand the Cluster**
** Add all neighboring points that are within ε distance of the core point.
** If a neighboring point also has enough MinPts neighbors, expand the cluster further.
** Keep repeating this until no more points can be added to the cluster.


* **Check for Outliers (Noise Points)**
** If a point never gets enough neighbors to become part of a cluster, it is marked as noise (outlier).
** Noise points are usually isolated or far from other dense regions.


* **Repeat Until All Points Are Processed**
** Pick another unvisited point and repeat the process.
** If it forms a new cluster, start expanding it.
** If not, move to the next point.
** Process ends when all points are either assigned to a cluster or marked as noise.

#### **K Means vs DBSCAN**
* **No need to choose K (clusters are formed based on density).**
* **Finds irregular shapes (K-Means only makes circular clusters).**
* **Ignores noise (K-Means forces every point into a cluster).**

### **PCA (Principal Component Analysis)**
PCA is a dimensionality reduction technique that simplifies data while keeping the most important information.

The goal is to reduce the number of features (dimensions) in a dataset while keeping its essential patterns.

* Standardize the data (mean = 0, variance = 1).  
* Compute the covariance matrix to find feature relationships.  
* Find eigenvectors & eigenvalues (new data axes).  
* Select top K eigenvectors (most important).  
* Project data onto new axes (dimensionality reduction)

pca12.png

pca11.png

* **Standardize the Data**
** Convert all features to the same scale (mean = 0, variance = 1).
** This ensures that no feature dominates because of different units (e.g., height in cm vs. weight in kg).

* **Compute the Covariance Matrix**
** A covariance matrix shows relationships between features.
** If two features are highly correlated, PCA will merge them into a principal component.

* **Find Principal Components (Eigenvectors & Eigenvalues)**
** Eigenvectors → New directions (axes) for the data.
** Eigenvalues → Importance (variance) of each eigenvector.
** The eigenvector with the highest eigenvalue becomes Principal Component 1 (PC1), the second highest becomes PC2, and so on.

**Think of it as finding a new set of axes that explain most of the data’s variance.**

* **Select the Top K Principal Components**
** Sort eigenvectors by eigenvalues (importance).
** Keep the top K components that capture the most variance.
** Drop the rest to reduce dimensions.

* **Transform the Data**
** The original dataset is projected onto the new principal components.
** The new dataset has fewer dimensions but retains the most important patterns.

### **t-SNE (t-Distributed Stochastic Neighbor Embedding)**

t-SNE is a dimensionality reduction algorithm mainly used for visualizing high-dimensional data in 2D or 3D.

* Compute similarities between data points in high dimensions.  
* Map similarities to 2D/3D space.  
* Adjust positions using gradient descent to preserve relationships.  

* Compute Pairwise Similarities in High-Dimensional Space
** t-SNE looks at each point and calculates how similar it is to other points.
** Uses a probability distribution where closer points have higher similarity.

* Map Similarities to a Lower-Dimensional Space
** The same similarity probabilities are calculated in low-dimensional space (2D or 3D).
** The goal is to keep relationships similar to the original space.

* Minimize the Difference (KL Divergence)
** It adjusts the new positions to make the low-dimensional similarities match the high-dimensional ones.
** This is done using gradient descent, continuously improving the placement.

## **Association Rule Learning-** Finding Relationships

* **Apriori Algorithm** – Finds frequent itemsets in transactions.
* **FP-Growth Algorithm** – Faster version of Apriori.

**Example: Market Basket Analysis (People who buy bread also buy butter).**

**Association Rule Learning** is used to find relationships between items in a dataset. It helps in predicting what items are likely to be bought together (market basket analysis).

The goal is to discover rules like "If a customer buys A, they are likely to buy B."

* Find frequent itemsets (Apriori, ECLAT, FP-Growth).  
* Generate rules like {A} → {B}.  
* Evaluate rules using Support, Confidence, and Lift.

* **Example:** Milk & Bread appear in 30% of transactions → It is a frequent itemset.

**Generate Association Rules**
* From frequent itemsets, create if-then rules like:
"If someone buys X, they will likely buy Y."

### **Apriori Algorithm**
Apriori is an algorithm used for finding patterns and relationships in large datasets, mainly in market basket analysis.

* Find frequent itemsets (based on support).  
* Generate association rules (based on confidence & lift).  
* Use rules for recommendations or predictions.

Apriori-Algorithm-2.png

* **Find Frequent Itemsets**
** Scan the dataset to find items that appear frequently together.
** Use a threshold called Support (minimum number of times an itemset appears).
** **Example:** If "Milk & Bread" appear in 30% of transactions, and Support is 20%, we keep this pair.

* **Generate Association Rules**
** Once frequent itemsets are found, create if-then rules like:
"If a customer buys A, they are likely to buy B"

**Example:**
* **Rule:** {Milk} → {Bread} (If someone buys Milk, they might buy Bread).
* **Confidence:** Measures how often the rule is true.
* **Lift:** Measures how strong the association is compared to random chance.
